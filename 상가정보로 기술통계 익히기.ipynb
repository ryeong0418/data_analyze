* shift+enter 키를 누르면 셀이 실행되고 커서가 다음 셀로 이동한다.
* enter키를 누르면 다시 편집상태로 돌아온다
* esc 키를 누르고
  * a키를 누르면 위에 셀이 추가됩니다.
  * b키를 누르면 아래에 셀이 추가됩니다.

  * m키를 누르면 문서 셀로 변경됩니다.
  * y키를 누르면 코드셀로 변경됩니다.
